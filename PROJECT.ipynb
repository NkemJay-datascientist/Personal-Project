```python
def pearson_correlation(x,y):
    if len(x)==len(y):
        sum_xy=sum((x-x.mean())*(y-y.mean()))
        sum_x_squared=sum((x-x.mean())**2)
        sum_y_square=sum((y-y.mean())**2)
        corr=sum_xy/np.sqrt(sum_x_square*sum_y_squared)
    return corr

# print(person_correlation(x,y))
# print(person_correlation(x,x))
```
